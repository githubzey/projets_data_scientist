```python
# On définit une fonction pour obtenir les variables 
def get_info_var2(variable):
    vari = []
    for i in range(20):
        if variable in data_c["hints"][i]["food"]:
            data_i = data_c['hints'][i]["food"][variable]
            vari.append(data_i)
        else:
            data_i = "unknown"
            vari.append(data_i)
    return vari
```

```python
# Liste des variables à extraire pour créer le DataFrame final
list_var = ["foodId", "label", "category", "foodContentsLabel", "image"]
products = pd.DataFrame({column: get_info_var2(column) for column in list_var})
products
```

# Les grands principes du RGPD
https://www.aurorebonavia-avocat.fr/grands-principes-rgpd/#:~:text=Principe%20RGPD%20n%C2%B01%20%3A%20lic%C3%A9it%C3%A9%2C%20loyaut%C3%A9%20et%20transparence&text=la%20loyaut%C3%A9%20%3A%20le%20traitement%20des,un%20traitement%20de%20ses%20donn%C3%A9es.

**Principe RGPD n°1 : licéité, loyauté et transparence** 

   Lorsqu’une entreprise traite des données personnelles, elle doit s’assurer de 3 fondements :
   * la licéité de la donnée : le traitement doit être licite si
   * la loyauté : le traitement des données doit correspondre à ce que l’entreprise a décrit à l’internaute concerné 
   * la transparence : rien ne doit être caché à la personne qui fait l’objet 
    
**Principe RGPD n°2 :  limitation des finalités**

   Chaque information personnelle récoltée doit être justifiée. Il est impossible de conserver les données sans          finalité concrète.
    
**Principe RGPD n°3 : minimisation des données**
    
Le responsable du traitement des données de l’entreprise ne doit traiter uniquement les données qui sont             nécessaires
    
**Principe RGPD n°4 : exactitude des données personnelles**

   En plus de leur pertinence, les données personnelles collectées se doivent d’être exactes
  
**Principe RGPD n°5 : limitation de la conservation**

   36 mois (3 ans) : c’est la durée maximum autorisée pour conserver les datas des personnes inactives dans votre base de données. Il vous est toutefois possible de les garder plus longtemps si vous les rendez anonymes ;
    
   13 mois : tous les 13 mois, vous devez redemander à vos visiteurs leur consentement  pour enregistrer leurs données via les cookies ;
    1 mois : si un utilisateur exerce son droit d’accès, vous avez 1 mois pour répondre à sa demande.
    
**Principe RGPD n°6 : intégrité et confidentialité**

   Le responsable du fichier des datas doit garantir la sécurité ainsi que la confidentialité des données dont il dispose.
